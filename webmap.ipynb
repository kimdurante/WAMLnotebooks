```python
import folium
```

```python
tTracts = f'https://opendata.arcgis.com/datasets/85a2e8e4bf994742a5855c1339517681_3.geojson'


style = lambda feature: {'fillColor': '#00ff00',
        'color': 'Green',
        'weight': 2}
                                  
m = folium.Map(
    location = [39.025, -120.032349], 
    tiles = 'Stamen Terrain', 
    zoom_start = 10
    )


folium.GeoJson(tTracts, 
               style_function = style
              ).add_to(m)


```
